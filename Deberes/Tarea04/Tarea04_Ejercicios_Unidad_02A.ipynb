1. 1. Use el método de la bisección para encontrar soluciones precisas dentro de 10^(-2) para $x^3 - 7x^2 + 14x - 6 = 0$ en cada intervalo.

a. [0, 1] 

b. [1, 3.2] 

c. [3.2, 4] 


```python
# ejercicio_biseccion.py
def f(x):
    return x**3 - 7*x**2 + 14*x - 6

def biseccion(a, b, tol=1e-2, max_iter=100):
    if f(a) * f(b) >= 0:
        raise ValueError("La función no cambia de signo en el intervalo dado.")
    
    iteracion = 0
    while (b - a) / 2 > tol and iteracion < max_iter:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iteracion += 1
    
    return (a + b) / 2

# Intervalos dados
intervalos = [(0, 1), (1, 3.2), (3.2, 4)]

# Encontrar y mostrar las raíces
for i, (a, b) in enumerate(intervalos):
    try:
        raiz = biseccion(a, b)
        print(f"Intervalo {chr(97 + i)} [{a}, {b}]: Raíz ≈ {raiz:.4f}")
    except ValueError as e:
        print(f"Intervalo {chr(97 + i)} [{a}, {b}]: {e}")
```

b. Use el método de bisección para encontrar una aproximación dentro de $10^{-3}$ para un valor en el intervalo $[-2, 0]$ tal que:

$$
x^2 - 1 = e^{1 - x^2}
$$

```python

import numpy as np

def f(x):
    return x**2 - 1 - np.exp(1 - x**2)

def biseccion(a, b, tol=1e-3, max_iter=100):
    if f(a) * f(b) >= 0:
        raise ValueError("No hay cambio de signo en el intervalo.")
    
    iteracion = 0
    while (b - a) / 2 > tol and iteracion < max_iter:
        c = (a + b) / 2
        if f(c) == 0:
            return c
        if f(a) * f(c) < 0:
            b = c
        else:
            a = c
        iteracion += 1
    
    return (a + b) / 2

# Aplicamos al intervalo [-2, 0] donde vemos que hay cambio de signo
try:
    raiz = biseccion(-2, 0)
    print(f"Solución encontrada en [-2, 0]: x ≈ {raiz:.5f}")
    print(f"Valor de la función en la raíz: f({raiz:.5f}) = {f(raiz):.2e}")
except ValueError as e:
    print(e)
```

5. Sea $f(x) = \dfrac{(x + 3)(x + 1)^2 x (x - 1)^3}{(x - 3)}$.

¿En qué cero de $f$ converge el método de bisección cuando se aplica en los siguientes intervalos?

a. $[-1.5,\ 2.5]$  
b. $[-0.5,\ 2.4]$  
c. $[-0.5,\ 3]$  
d. $[-3,\ -0.5]$


```python
def f(x):
    return (x + 3) * (x + 1)**2 * x * (x - 1)**3 * (x - 3)

# Evaluar en los intervalos
intervalos = [
    ([-1.5, 2.5], "a. [-1.5, 2.5]"),
    ([-0.5, 2.4], "b. [-0.5, 2.4]"),
    ([-0.5, 3], "c. [-0.5, 3]"),
    ([-3, -0.5], "d. [-3, -0.5]")
]

for intervalo, nombre in intervalos:
    a, b = intervalo
    fa, fb = f(a), f(b)
    print(f"{nombre}: f({a}) = {fa:.2f}, f({b}) = {fb:.2f}")
    if fa * fb < 0:
        print("  → Hay cambio de signo. Converge a una raíz.")
    elif fa == 0:
        print(f"  → Converge exactamente a x = {a} (f(x) = 0).")
    elif fb == 0:
        print(f"  → Converge exactamente a x = {b} (f(x) = 0).")
    else:
        print("  → No hay cambio de signo. No converge.")
```

1. Un abrevadero de longitud $L$ tiene una sección transversal en forma de semicírculo con radio $r$.  
(Consulte la figura adjunta.)  
Cuando se llena con agua hasta una distancia $h$ a partir de la parte superior, el volumen $V$ de agua es:

$$
V = L \left[ 0.5 \pi r^2 - r^2 \arcsin\left(\frac{h}{r}\right) - h \sqrt{r^2 - h^2} \right]
$$

![alt text](../images/cap1.png)

Suponga que $L = 10\ \text{cm}$, $r = 1\ \text{cm}$ y $V = 12.4\ \text{cm}^3$.  
Encuentre la profundidad del agua en el abrevadero con una precisión de $0.01\ \text{cm}$.


```python
import math

def calcular_profundidad_agua(L, r, V):
      
    def volumen_formula(h):
        if h >= r:
            # Si h >= r, el agua llena toda la sección semicircular
            return L * (0.5 * math.pi * r**2)
        
        # Términos de la fórmula
        termino1 = 0.5 * math.pi * r**2
        termino2 = r**2 * math.asin(h/r)
        termino3 = h * math.sqrt(r**2 - h**2)
        
        volumen = L * (termino1 - termino2 - termino3)
        return volumen
    
    # Método de bisección para encontrar h
    h_min = 0
    h_max = r
    tolerancia = 1e-6
    
    while h_max - h_min > tolerancia:
        h_medio = (h_min + h_max) / 2
        volumen_calculado = volumen_formula(h_medio)
        
        if volumen_calculado < V:
            h_min = h_medio
        else:
            h_max = h_medio
    
    return (h_min + h_max) / 2

# Datos del problema
L = 10  # cm
r = 1   # cm
V = 12.4  # cm³

# Calcular la profundidad
h = calcular_profundidad_agua(L, r, V)

print(f"Datos del problema:")
print(f"Longitud del abrevadero (L): {L} cm")
print(f"Radio de la sección (r): {r} cm")
print(f"Volumen de agua (V): {V} cm³")
print(f"\nProfundidad del agua (h): {h:.3f} cm")

# Verificación del resultado
def verificar_resultado(L, r, h):
    termino1 = 0.5 * math.pi * r**2
    termino2 = r**2 * math.asin(h/r)
    termino3 = h * math.sqrt(r**2 - h**2)
    volumen_verificado = L * (termino1 - termino2 - termino3)
    return volumen_verificado

V_verificado = verificar_resultado(L, r, h)
print(f"\nVerificación:")
print(f"Volumen calculado con h = {h:.3f} cm: {V_verificado:.3f} cm³")
print(f"Error: {abs(V - V_verificado):.6f} cm³")

# Convertir a la precisión solicitada (0.01 cm)
h_redondeado = round(h, 2)
print(f"\nRespuesta final (redondeada a 0.01 cm): {h_redondeado} cm")
```

2. Un objeto que cae verticalmente a través del aire está sujeto a una resistencia viscosa, así como a la fuerza de gravedad.  
Suponga que un objeto con masa $m$ cae desde una altura $s_0$ y que la altura del objeto después de $t$ segundos está dada por:

$$
s(t) = s_0 - \frac{mg}{k}t + \frac{m^2 g}{k^2} \left( 1 - e^{-\frac{kt}{m}} \right),
$$

donde $g = 9.81\ \mathrm{m/s^2}$ y $k$ representa el coeficiente de resistencia del aire en $\mathrm{N \cdot s/m}$.

Suponga:  
- $s_0 = 300\ \mathrm{m}$  
- $m = 0.25\ \mathrm{kg}$  
- $k = 0.1\ \mathrm{N \cdot s/m}$

Encuentre, con una precisión de $0.01$ segundos, el tiempo que tarda el objeto en golpear el piso (es decir, cuando $s(t) = 0$).


```python
import math

def calcular_tiempo_caida(s0, m, k, g):
    
    def posicion(t):
        # Ecuación dada: s(t) = s0 - (mg/k)t + (m²g/k²)(1 - e^(-kt/m))
        termino1 = s0
        termino2 = -(m * g / k) * t
        termino3 = (m**2 * g / k**2) * (1 - math.exp(-k * t / m))
        return termino1 + termino2 + termino3
    
    # Método de bisección para encontrar cuando s(t) = 0
    t_min = 0
    t_max = 100  # Tiempo máximo estimado (segundos)
    tolerancia = 1e-6
    
    # Verificar que hay cambio de signo
    while posicion(t_max) > 0:
        t_max *= 2
    
    # Bisección
    while t_max - t_min > tolerancia:
        t_medio = (t_min + t_max) / 2
        if posicion(t_medio) > 0:
            t_min = t_medio
        else:
            t_max = t_medio
    
    return (t_min + t_max) / 2

# Datos del problema
s0 = 300    # altura inicial en metros
m = 0.25    # masa en kg
k = 0.1     # coeficiente de resistencia en Ns/m
g = 9.81    # aceleración de la gravedad en m/s²

print("PROBLEMA DE CAÍDA CON RESISTENCIA DEL AIRE")
print("="*50)
print(f"Datos del problema:")
print(f"Altura inicial (s0): {s0} m")
print(f"Masa del objeto (m): {m} kg")
print(f"Coeficiente de resistencia (k): {k} Ns/m")
print(f"Gravedad (g): {g} m/s²")
print()

# Calcular velocidad terminal
v_terminal = (m * g) / k
print(f"Velocidad terminal: v_t = mg/k = {v_terminal:.2f} m/s")
print()

# Calcular el tiempo de caída
t_caida = calcular_tiempo_caida(s0, m, k, g)

print(f"Tiempo de caída: {t_caida:.6f} segundos")
print(f"Tiempo redondeado (0.01 s): {round(t_caida, 2)} segundos")

# Verificación del resultado
def posicion_verificacion(t):
    termino1 = s0
    termino2 = -(m * g / k) * t
    termino3 = (m**2 * g / k**2) * (1 - math.exp(-k * t / m))
    return termino1 + termino2 + termino3

s_final = posicion_verificacion(t_caida)
print(f"\nVerificación:")
print(f"Posición final s({t_caida:.6f}) = {s_final:.8f} m")
print(f"(Debería ser aproximadamente 0)")

# Calcular velocidad final
def velocidad(t):
    # v(t) = ds/dt = -(mg/k) + (mg/k)e^(-kt/m)
    return -(m * g / k) + (m * g / k) * math.exp(-k * t / m)

v_final = velocidad(t_caida)
print(f"\nVelocidad final: {v_final:.2f} m/s")
print(f"Porcentaje de velocidad terminal: {(abs(v_final)/v_terminal)*100:.1f}%")

# Crear tabla de posición vs tiempo para visualizar la caída
print(f"\nTabla de caída:")
print("Tiempo (s) | Posición (m) | Velocidad (m/s)")
print("-" * 40)
for i in range(0, int(t_caida) + 10, 5):
    if i <= t_caida:
        pos = posicion_verificacion(i)
        vel = velocidad(i)
        print(f"{i:8.1f} | {pos:10.2f} | {vel:12.2f}")
    if pos <= 0:
        break
```

**1.** Use el **Teorema 2.1** para encontrar una cota para el número de iteraciones necesarias para lograr una aproximación con precisión de $10^{-4}$ para la solución de:

$$
x^3 - x - 1 = 0
$$

que se encuentra dentro del intervalo $[1,\ 2]$.

Luego, encuentre una aproximación para la raíz con este grado de precisión.


```python
def ecuacion_c(x):
    return x**3 - x - 1

# Definimos el intervalo y los parámetros
a, b = 1, 2
TOL = 10**-4
N0 = 14

# Lista para almacenar los puntos medios
puntos_medios = []

# Ejecutamos el método de bisección y almacenamos los puntos medios
def biseccionTeorico(ecuacion, a, b, TOL, N0):
    FA = ecuacion(a)
    p = a

    for i in range(1, N0 + 1):
        p_anterior = p
        p = (a + b) / 2
        FP = ecuacion(p)
        puntos_medios.append(p)  # Guardamos el punto medio externo a la función

        if FP == 0 or abs(p - p_anterior) < TOL:
            return p

        if FA * FP > 0:
            a = p
            FA = FP
        else:
            b = p

    return None

# Ejecutamos el método de bisección
raiz_aproximada = biseccionTeorico(ecuacion_c, a, b, TOL, N0)

# Generamos la gráfica
x_values = np.linspace(0, 2.5, 500)
y_values = ecuacion_c(x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="f(x) = x³ - x - 1", color="blue")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(raiz_aproximada, color="green", linestyle="--", label=f"Raíz aproximada: x ≈ {raiz_aproximada:.5f}")


plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Método de Bisección aplicado a f(x) = x³ - x - 1")
plt.legend()
plt.grid(True)
plt.show()
```

2. La función definida por $f(x) = \sin(\pi x)$ tiene ceros en cada número entero.  
Muestre que cuando $-1 < a < 0$ y $2 < b < 3$, el método de bisección converge a:

a. $0$, si $a + b < 2$  
b. $2$, si $a + b > 2$  
c. $1$, si $a + b = 2$


```python
from math import sin, pi

def bisec(a, b, tol=1e-10):
    f = lambda x: sin(pi * x)
    while abs(b - a) > tol:
        m = (a + b) / 2
        if f(a) * f(m) <= 0: b = m
        else: a = m
    return round((a + b) / 2)

# Ejemplos
print(bisec(-0.5, 2.3))  # a + b < 2 => 0
print(bisec(-0.5, 2.6))  # a + b > 2 => 2
print(bisec(-0.5, 2.5))  # a + b = 2 => 1
```
