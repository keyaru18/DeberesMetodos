# Conjunto de Ejercicios

Determine el orden de la mejor aproximación para las siguientes funciones, usando la Serie de Taylor y el Polinomio de Lagrange:

1. $\frac{1}{25*x^2+1}$, $x_0 = 0$
2. $arctanx$, $x_0 = 1$

- Escriba las formulas de los diferentes polinomios
- Grafique las diferentes aproximaciones

**Serie de Tailor**

$P_n(x) = f(x_0) + f'(x_0)(x - x_0) + \frac{f(x_0)}{2!}(x - x_0)^2 + ... + \frac{f^{(n)}(x_0)}{n!}(x - x_0)^n$

$\sum_{k=0}^{n} \frac{f^{(k)}(x_0)}{k!}(x - x_0)^k$

**Polinomio de Lagrange**

$P(x) = f(x_0)L_0(x) + ... + f(x_n)L_n(x)$

$L_k(x) = \frac{(x - x_0)(x - x_1) ... (x - x_{k-1})(x - x_{k+1})...(x - x_n)}{(x_k - x_0)(x_k - x_1) ... (x_k - x_{k-1})(x_k - x_{k+1}) ... (x_k - x_n)}$

$ = \prod_{\substack{i=0 \\ i \ne k}}^{n} \frac{(x - x_i)}{(x_k - x_i)}$

## 1. $\frac{1}{25*x^2+1}$, $x_0 = 0$
- Serie de Taylor

[Repositorio de Github](https://github.com/keyaru18/DeberesMetodos.git)

```python
import sympy as sp

# Definir la variable y la función
x = sp.symbols('x')
f = 1 / (25 * x**2 + 1)

# Calcular las derivadas
f_prime = sp.diff(f, x)
f_double_prime = sp.diff(f_prime, x)

# Evaluar las derivadas en x0 = 0
x0 = 0
f0 = f.subs(x, x0)
f_prime0 = f_prime.subs(x, x0)
f_double_prime0 = f_double_prime.subs(x, x0)

# Construir el polinomio de Taylor de grado 2
taylor_poly = f0 + f_prime0 * x + (f_double_prime0 / 2) * x**2

# Mostrar el resultado
print("El polinomio de Taylor de grado 2 es:")
sp.pprint(taylor_poly)
```

- Gráfica de la aproximación

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función original y el polinomio de Taylor
def f_original(x):
    return 1 / (25 * x**2 + 1)

def taylor_approx(x):
    return 1 - 25 * x**2

# Rango de valores para x (centrado en x0=0)
x_vals = np.linspace(-0.5, 0.5, 400)  # Ajusta el rango según necesidad

# Calcular los valores de las funciones
y_original = f_original(x_vals)
y_approx = taylor_approx(x_vals)

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_original, label='Función original: $\dfrac{1}{25x^2 + 1}$', color='blue')
plt.plot(x_vals, y_approx, label='Aproximación de Taylor: $1 - 25x^2$', color='red', linestyle='--')
plt.title('Comparación: Función original vs. Aproximación de Taylor (grado 2)', fontsize=12)
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

- Polinomio de Lagrange

```python
import numpy as np

# Función original
def f(x):
    return 1 / (25 * x**2 + 1)

# Puntos de interpolación (ejemplo para grado 2: 3 puntos)
x_points = np.array([-0.2, 0, 0.2])  # Puntos alrededor de x0=0
y_points = f(x_points)

print("Puntos de interpolación (x, y):")
for xi, yi in zip(x_points, y_points):
    print(f"({xi}, {yi:.4f})")
```

```python
from scipy.interpolate import lagrange

# Construir el polinomio de Lagrange
poly_lagrange = lagrange(x_points, y_points)

# Mostrar el polinomio (coeficientes)
print("\nPolinomio de Lagrange (coeficientes):")
print(poly_lagrange)
```

- Gráfica de la aproximación

```python
import matplotlib.pyplot as plt

# Rango para graficar
x_vals = np.linspace(-0.5, 0.5, 400)
y_original = f(x_vals)
y_lagrange = poly_lagrange(x_vals)

# Gráfica
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_original, label='Función original', color='blue')
plt.plot(x_vals, y_lagrange, label='Polinomio de Lagrange', color='red', linestyle='--')
plt.scatter(x_points, y_points, color='green', label='Puntos de interpolación')
plt.title('Aproximación con Polinomio de Lagrange (grado 2)', fontsize=12)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

## 2. $arctanx$, $x_0 = 1$
- Serie de Taylor

```python
import numpy as np
from sympy import symbols, atan, diff, simplify
import math  # Usamos math.factorial en lugar de np.math.factorial

# Definir símbolo y función
x = symbols('x')
f = atan(x)
x0 = 1

# Calcular derivadas y evaluar en x0
print("=== Derivadas de arctan(x) evaluadas en x₀ = 1 ===")
derivadas = []
for n in range(4):  # Hasta la 3ra derivada
    derivada = diff(f, x, n)
    valor = derivada.subs(x, x0)
    derivadas.append((derivada, valor))
    print(f"f^({n})(x) = {derivada}")
    print(f"f^({n})(1) = {valor}\n")

# Construir polinomio de Taylor
P3 = sum(val * (x - x0)**k / math.factorial(k) for k, (_, val) in enumerate(derivadas))
P3_simplificado = simplify(P3)

print("\n=== Polinomio de Taylor de grado 3 ===")
print(f"P3(x) = {P3_simplificado}")
```

- Gráfica de la aproximación

```python
import matplotlib.pyplot as plt
from sympy import lambdify

%matplotlib inline

# Convertir a funciones numéricas
f_num = lambdify(x, f, 'numpy')
P3_num = lambdify(x, P3_simplificado, 'numpy')

# Graficar
xx = np.linspace(0, 2, 400)
plt.figure(figsize=(10, 5))
plt.plot(xx, f_num(xx), label='arctan(x)', linewidth=2)
plt.plot(xx, P3_num(xx), '--', label='Aproximación P3(x)', linewidth=2)
plt.scatter([x0], [f_num(x0)], color='red', s=100, label=f'Punto x₀={x0}')
plt.title('Aproximación de Taylor para arctan(x) alrededor de x₀=1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

- Polinomio de Lagrange

```python
import numpy as np
from sympy import symbols, atan, simplify, expand
import math

# Definir símbolo y función
x = symbols('x')
f = atan(x)

# Puntos de interpolación alrededor de x₀=1
x_values = [0.5, 0.75, 1.0, 1.25, 1.5]
y_values = [atan(xi).evalf() for xi in x_values]

# Construir polinomio de Lagrange
def lagrange_poly(x_val, x_data, y_data):
    n = len(x_data)
    poly = 0
    for i in range(n):
        term = y_data[i]
        for j in range(n):
            if j != i:
                term *= (x_val - x_data[j])/(x_data[i] - x_data[j])
        poly += term
    return poly

L4 = lagrange_poly(x, x_values, y_values)
L4_simplified = expand(simplify(L4))

print("=== Puntos de interpolación ===")
for xi, yi in zip(x_values, y_values):
    print(f"f({xi:.2f}) = {yi:.4f}")

print("\n=== Polinomio de Lagrange de grado 4 ===")
print(f"L4(x) = {L4_simplified}")
```

- Gráfica de la aproximación

```python
import matplotlib.pyplot as plt
from sympy import lambdify

%matplotlib inline

# Convertir a funciones numéricas
f_num = lambdify(x, f, 'numpy')
L4_num = lambdify(x, L4_simplified, 'numpy')

# Graficar
xx = np.linspace(0.4, 1.6, 400)
plt.figure(figsize=(10, 6))

# Función original y aproximación
plt.plot(xx, f_num(xx), label='arctan(x)', linewidth=2)
plt.plot(xx, L4_num(xx), '--', label='Polinomio Lagrange L4(x)', linewidth=2)

# Puntos de interpolación
plt.scatter(x_values, y_values, color='red', s=100, label='Puntos de interpolación')

plt.title('Aproximación con Polinomio de Lagrange para arctan(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
