1. Dados los puntos (0,1), (1,5), (2,3), determine el *spline cúbico*.

- **$S(x)$** es un spline cúbico  
  - $S_j(x) = a_j + b_j(x - x_j) + c_j(x - x_j)^2 + d_j(x - x_j)^3$

$S_0(0) = 1$: $a_0 = 1$

$S_0(1) = 5$: $a_0 + b_0 + c_0 + d_0 = 5$

$S_1(1) = 5$: $a_1 = 5$

$S_1(2) = 3$: $a_1 + b_1 + c_1 + d_1 = 3$

Las derivadas de los splines son:

$S_0'(x) = b_0 + 2c_0 x + 3d_0 x^2$

$S_0''(x) = 2c_0 + 6d_0 x$

$S_1'(x) = b_1 + 2c_1(x - 1) + 3d_1(x - 1)^2$

$S_1''(x) = 2c_1 + 6d_1(x - 1)$

En $x = 1$:

$S_0'(1) = S_1'(1) \Rightarrow b_0 + 2c_0 + 3d_0 = b_1$

$S_0''(1) = S_1''(1) \Rightarrow 2c_0 + 6d_0 = 2c_1$

a) Frontera natural:

1. $a_0 = 1$ y $a_1 = 5$.

2. $a_0 + b_0 + c_0 + d_0 = 5$ y $c_0 = 0$:

$$1 + b_0 + d_0 = 5 \implies b_0 + d_0 = 4$$

3. $a_1 + b_1 + c_1 + d_1 = 3$:

$$5 + b_1 + c_1 + d_1 = 3 \implies b_1 + c_1 + d_1 = -2$$

4. $b_0 + 2 c_0 + 3 d_0 = b_1$ y $c_0 = 0$:

$$b_0 + 3 d_0 = b_1$$

5. $2 c_0 + 6 d_0 = 2 c_1$ y $c_0 = 0$:

$$6 d_0 = 2 c_1 \implies c_1 = 3 d_0$$

6. $2 c_1 + 6 d_1 = 0$:

$$2 (3 d_0) + 6 d_1 = 0 \implies 6 d_0 + 6 d_1 = 0 \implies d_0 = - d_1$$

Sustituyendo $d_1 = - d_0$ en $b_1 + c_1 + d_1 = -2$:

$$b_1 + 3 d_0 - d_0 = -2 \implies b_1 + 2 d_0 = -2$$

$b_1 = b_0 + 3 d_0$ y $b_0 = 4 - d_0$:

$$(4 - d_0) + 3 d_0 + 2 d_0 = -2 \implies 4 + 4 d_0 = -2 \implies d_0 = -\frac{3}{2}$$

Entonces:

$b_0 = 4 - d_0 = 4 - \left(-\frac{3}{2}\right) = \frac{11}{2}$

$b_1 = b_0 + 3 d_0 = \frac{11}{2} + 3 \left(-\frac{3}{2}\right) = \frac{11}{2} - \frac{9}{2} = 1$

$c_1 = 3 d_0 = 3 \left(-\frac{3}{2}\right) = -\frac{9}{2}$

$d_1 = - d_0 = \frac{3}{2}$

**Spline cúbico con frontera natural:**

$S_0(x) = 1 + \frac{11}{2} x - \frac{3}{2} x^3$ para $x \in [0,1]$

$S_1(x) = 5 + (x - 1) - \frac{9}{2} (x - 1)^2 + \frac{3}{2} (x - 1)^3$ para $x \in [1,2]$

b) Frontera condicionada:

$a_0 = 1$ y $a_1 = 5$.

$b_0 = 2$.

$a_0 + b_0 + c_0 + d_0 = 5$:

$1 + 2 + c_0 + d_0 = 5 \implies c_0 + d_0 = 2$

$2c_0 + 6d_0 = 2c_1$:

$(2 + 2c_0 + 3d_0) + 2(c_0 + 3d_0) + 3d_1 = -1$

Simplificando:

$2 + 4c_0 + 9d_0 + 3d_1 = -1 \implies 4c_0 + 9d_0 + 3d_1 = -3$

$c_0 + d_0 = 2$ y $c_1 = c_0 + 3d_0$, y $b_1 + c_1 + d_1 = -2$:

$(2 + 2c_0 + 3d_0) + (c_0 + 3d_0) + d_1 = -2 \implies 2 + 3c_0 + 6d_0 + d_1 = -2$

$3c_0 + 6d_0 + d_1 = -4$

Resolviendo el sistema:

$c_0 = 2 - d_0$

Sustituyendo en $3c_0 + 6d_0 + d_1 = -4$:

$3(2 - d_0) + 6d_0 + d_1 = -4 \implies 6 + 3d_0 + d_1 = -4 \implies 3d_0 + d_1 = -10$

Sustituyendo en $4c_0 + 9d_0 + 3d_1 = -3$:

$4(2 - d_0) + 9d_0 + 3d_1 = -3 \implies 8 + 5d_0 + 3d_1 = -3 \implies 5d_0 + 3d_1 = -11$

Resolviendo:

\[
\begin{cases}

$9d_0 + 3d_1 = -30$

$S_1(x) = 5 + \frac{5}{4}(x - 1) - \frac{15}{2}(x - 1)^2 + \frac{17}{4}(x - 1)^3 \quad \text{para } x \in [1, 2]$

2. Dados los puntos $(-1, 1)$ y $(1, 3)$, determine el spline cúbico sabiendo que:
$f'(x_0) = 1$ y $f'(x_n) = 2$.

- Interpolación en los puntos:

    $S(-1) = 1 \implies a = 1$

    $S(1) = 3 \implies a + 2b + 4c + 8d = 3$

- Derivadas en los extremos:

    $S'(x) = b + 2c(x - x_0) + 3d(x - x_0)^2$

    $S'(-1) = 1 \implies b = 1$

    $S'(1) = 2 \implies b + 4c + 12d = 2$

Sustituyendo $a = 1$ y $b = 1$ en las ecuaciones:

1. $S(1) = 3$:
   $$
   1 + 2(1) + 4c + 8d = 3 \implies 4c + 8d = 0 \quad \text{(1)}
   $$

2. $S'(1) = 2$:
   $$
   1 + 4c + 12d = 2 \implies 4c + 12d = 1 \quad \text{(2)}
   $$

Restando (1) de (2):
$$
(4c + 12d) - (4c + 8d) = 1 - 0 \implies 4d = 1 \implies d = \frac{1}{4}
$$

Sustituyendo $d$ en (1):
$$
4c + 8\left(\frac{1}{4}\right) = 0 \implies 4c + 2 = 0 \implies c = -\frac{1}{2}
$$

**Spline cúbico final**
$$
S(x) = 1 + (x + 1) - \frac{1}{2}(x + 1)^2 + \frac{1}{4}(x + 1)^3
$$

Forma simplificada:
$$
S(x) = \frac{1}{4}x^3 - \frac{1}{2}x^2 + \frac{5}{4}x + \frac{5}{4}
$$

3. Diríjase al pseudocódigo del spline cúbico con frontera natural provisto en clase, en
base a ese pseudocódigo complete la siguiente función:




```python
import sympy as sym

def cubic_spline(xs: list[float], ys: list[float]) -> list[sym.Symbol]:
    """
    Cubic spline interpolation ``S``. Every two points are interpolated by a cubic polynomial
    ``S_j`` of the form ``S_j(x) = a_j + b_j(x - x_j) + c_j(x - x_j)^2 + d_j(x - x_j)^3.``

    Parameters:
    - xs, ys: points to be interpolated (lists of equal length)

    Return:
    - List of symbolic expressions for each segment of the cubic spline
    """
    # Sort points by x-coordinate
    points = sorted(zip(xs, ys), key=lambda x: x[0])
    xs = [x for x, _ in points]
    ys = [y for _, y in points]
    
    n = len(xs) - 1  # number of spline segments
    
    if n < 1:
        raise ValueError("At least two points are needed for interpolation")
    
    h = [xs[i+1] - xs[i] for i in range(n)]  # differences between x-values
    
    # Step 1: Initialize alpha (corrected size to n+1)
    alpha = [0] * (n + 1)
    for i in range(1, n):
        alpha[i] = 3/h[i]*(ys[i+1]-ys[i]) - 3/h[i-1]*(ys[i]-ys[i-1])
    
    # Step 2: Solve the tridiagonal system for c
    # Thomas algorithm implementation
    l = [1]  # l_0 = 1
    u = [0]   # u_0 = 0
    z = [0]   # z_0 = 0
    
    for i in range(1, n):
        l.append(2*(xs[i+1]-xs[i-1]) - h[i-1]*u[i-1])
        u.append(h[i]/l[i])
        z.append((alpha[i]-h[i-1]*z[i-1])/l[i])
    
    l.append(1)  # l_n = 1
    z.append(0)  # z_n = 0
    c = [0] * (n + 1)  # c_n = 0
    
    # Back substitution
    for j in range(n-1, -1, -1):
        c[j] = z[j] - u[j]*c[j+1]
    
    # Step 3: Compute the splines
    x = sym.Symbol('x')
    splines = []
    
    for j in range(n):
        b = (ys[j+1]-ys[j])/h[j] - h[j]*(c[j+1]+2*c[j])/3
        d = (c[j+1]-c[j])/(3*h[j])
        S = ys[j] + b*(x-xs[j]) + c[j]*(x-xs[j])**2 + d*(x-xs[j])**3
        splines.append(S)
    
    return splines
```

```python
xs = [0, 1, 2]
ys = [-5, -4, 3]

splines = cubic_spline(xs=xs, ys=ys)

for i, s in enumerate(splines):
    print(f"S_{i}(x) = {s}")
    print(f"Simplificado: {s.expand()}\n")
```

4. Use la función `cubic_spline_clamped`, provista en el enlace de Github, para graficar
los datos de la siguiente tabla.



```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

def cubic_spline_clamped(xs, ys, fpo, fpn):
    """
    Computes the clamped cubic spline interpolation with specified first derivatives at endpoints.
    
    Parameters:
    xs, ys: points to be interpolated
    fpo: first derivative at x0 (f'(x0))
    fpn: first derivative at xn (f'(xn))
    
    Returns:
    List of symbolic expressions for each segment of the cubic spline
    """
    # Sort points by x-coordinate
    points = sorted(zip(xs, ys), key=lambda x: x[0])
    xs = [x for x, _ in points]
    ys = [y for _, y in points]
    
    n = len(xs) - 1
    
    if n == 0:
        raise ValueError("At least two points are needed for interpolation")
    
    h = [xs[i+1] - xs[i] for i in range(n)]
    
    # Step 1: Initialize alpha (different for clamped spline)
    alpha = [0] * (n + 1)
    alpha[0] = 3*(ys[1]-ys[0])/h[0] - 3*fpo
    alpha[n] = 3*fpn - 3*(ys[n]-ys[n-1])/h[n-1]
    
    for i in range(1, n):
        alpha[i] = 3/h[i]*(ys[i+1]-ys[i]) - 3/h[i-1]*(ys[i]-ys[i-1])
    
    # Step 2: Solve the tridiagonal system (different for clamped)
    l = [2*h[0]] + [0]*n
    mu = [0.5] + [0]*n
    z = [alpha[0]/l[0]] + [0]*n
    
    for i in range(1, n):
        l[i] = 2*(xs[i+1]-xs[i-1]) - h[i-1]*mu[i-1]
        mu[i] = h[i]/l[i]
        z[i] = (alpha[i]-h[i-1]*z[i-1])/l[i]
    
    l[n] = h[n-1]*(2-mu[n-1])
    z[n] = (alpha[n]-h[n-1]*z[n-1])/l[n]
    
    c = [0]*(n+1)
    c[n] = z[n]
    
    for i in range(n-1, -1, -1):
        c[i] = z[i] - mu[i]*c[i+1]
    
    # Step 3: Compute b and d coefficients
    b = [0]*n
    d = [0]*n
    
    for i in range(n):
        b[i] = (ys[i+1]-ys[i])/h[i] - h[i]*(c[i+1]+2*c[i])/3
        d[i] = (c[i+1]-c[i])/(3*h[i])
    
    # Step 4: Create symbolic expressions
    x = sym.Symbol('x')
    splines = []
    
    for i in range(n):
        expr = ys[i] + b[i]*(x-xs[i]) + c[i]*(x-xs[i])**2 + d[i]*(x-xs[i])**3
        splines.append(expr)
    
    return splines, xs
```

- Graficas de splines

```python
def plot_splines(splines, xs, ys, title="Cubic Spline Interpolation"):
    """
    Plots the cubic spline interpolation along with the original data points.
    """
    x = sym.Symbol('x')
    x_vals = np.linspace(min(xs), max(xs), 500)
    
    plt.figure(figsize=(10, 6))
    
    # Plot each spline segment
    for i in range(len(splines)):
        x_start = xs[i]
        x_end = xs[i+1]
        x_segment = np.linspace(x_start, x_end, 100)
        
        # Convert sympy expression to numpy function
        f = sym.lambdify(x, splines[i], 'numpy')
        y_segment = f(x_segment)
        
        plt.plot(x_segment, y_segment, 'b-')
    
    # Plot original data points
    plt.plot(xs, ys, 'ro', label='Data points')
    
    plt.title(title)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.legend()
    plt.show()
```
